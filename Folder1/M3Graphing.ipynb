```python
# Allow graphs to appear in Jupyter notebook
%matplotlib inline

# import packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
from IPython.display import display

# read data and assign to dataframes
df_Constrn = pd.read_csv('ConstructionTimeSeriesDatav2.csv')

#Display each dataframe heads
display(df_Constrn.head())
```

```python
# reformat Month-Year column in df
df_Constrn['Month-Year'] = [i.replace("-", " '") for i in df_Constrn['Month-Year']]
```

```python
# define x,y values for plot
x = df_Constrn['Month-Year']
y1 = df_Constrn['Private Construction']/1000
y2 = df_Constrn['Public Construction']/1000
```

```python
# create plot
fig,ax = plt.subplots()
ax.plot(x,y1, color='b', label='Private Construction', linewidth='2')
ax.plot(x,y2, color='olive', label='Public Construction', linewidth='2')

# Remove top,right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Set axes labels & ticks
ax.set_xlabel('Month',fontsize='14',fontname='Times New Roman')
ax.set_ylabel('Spending (Billions USD)',fontsize='14',fontname='Times New Roman')
ax.set_xticks(x[::24])
ax.set_xticklabels(x[::24], fontname='Times New Roman', fontsize='11')
ax.set_ylim(0,max(y1)+10)
for tick in ax.get_yticklabels():
    tick.set_fontname('Times New Roman'), tick.set_fontsize('11')
ax.yaxis.set_major_formatter(tck.StrMethodFormatter('${x}'))

# highlight aspect of interest
ax.axvline(x=71,lw=.75,color='k',label='Recession begins',linestyle='--')
ax.axvline(x=90,lw=.75,color='k',label='Recession ends',linestyle=':')
ax.axvspan(71,90,ymin=0,ymax = 300,facecolor='r',alpha=0.1)
ax.text(93,67,'2007-2009\nGreat Recession',fontsize=12,verticalalignment='center',
        color='r', horizontalalignment='left',fontname='Times New Roman')
plt.plot([85, 92], [70, 69], color='r', lw=.5)

# add legend
ax.legend(loc=1, prop={'size': 8})

# Chart titles
fig.suptitle('Comparison of U.S. Private vs Public Construction Spending',
            fontname='Times New Roman', fontsize='18')   # Graph title
plt.title('2002-2014', fontsize='16',loc='Center',fontname='Times New Roman')
```

The aspect of interest, portrayed by the light-red span and bordered by two dashed lines, identifies the date range of Dec 2007 to June 2009. This range corresponds with the timeline of the 2007-2009 "Great Recession".  During this time, Americans faced the subprime mortgage crisis and the housing bubble burst. The highlighted region in the plot shows an intruiging trend: despite the significant decline of U.S. private construction spending during and after the Great Recession, public construction spending remained constant and, in fact, appears to have increased slightly.
